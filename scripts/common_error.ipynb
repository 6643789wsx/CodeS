```python
import matplotlib.pyplot as plt

font_size = 13

# Data
errors = ['ImportError', 'NotImplementedError', 'SyntaxError']
code_sextended_values = [12, 16, 24]
code_sextended_pt_values = [17, 18, 25]
github_copilot_values = [21, 19, 21]

# X locations for the groups
ind = range(len(errors))
width = 0.25

fig, ax = plt.subplots()
rects1 = ax.bar([p - width for p in ind], code_sextended_values, width, label='CodeS Exteneded (SFT)',
                color='#76c7c0', hatch='/')
rects2 = ax.bar(ind, code_sextended_pt_values, width, label='CodeS Exteneded (PT)',
                color='#ff9999', hatch='x') 
rects3 = ax.bar([p + width for p in ind], github_copilot_values, width, label='GitHub Copilot',
                color='#f9d057', hatch='\\')

ax.set_ylabel('Occurrences', fontsize=font_size)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.tick_params(axis='y', labelsize=font_size)

# plt.xticks(rotation=8, fontsize=font_size)

ax.grid(True, which='both', linestyle='--', linewidth=0.5)

ax.set_xticks([p for p in ind])  # 使用原始索引位置使标签居中
ax.set_xticklabels(errors, fontsize=font_size)
ax.legend(fontsize=font_size)

# plt.show()
plt.savefig('./common_errors.pdf', bbox_inches='tight', format='pdf', dpi=600)
```
